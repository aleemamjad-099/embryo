```python
import cv2
from ultralytics import YOLO
import numpy as np

# 1. Model load karein
model = YOLO('best.pt') 

# 2. Image scan karein
results = model.predict('20230304-17097-3-F15-157.jpg', conf=0.5)
boxes = results[0].boxes

strongest_cell_idx = -1
max_score = 0

print("--- Cell Strength Analysis ---")

for i, box in enumerate(boxes):
    # Confidence Score nikaalein (YOLO kitna sure hai)
    conf = float(box.conf[0])
    
    # Size nikaalein (Width aur Height)
    x1, y1, x2, y2 = box.xyxy[0]
    w = x2 - x1
    h = y2 - y1
    
    # Aspect Ratio (Symmetry check): Agar 1.0 ke qareeb hai to matlab perfect round hai
    aspect_ratio = min(w, h) / max(w, h)
    
    # Final Health Score Formula (Hum khud bana rahe hain)
    # Score = Confidence + Symmetry
    health_score = (conf * 0.6) + (float(aspect_ratio) * 0.4)
    
    print(f"Cell #{i+1}: Confidence: {conf:.2f}, Symmetry: {aspect_ratio:.2f}, Health Score: {health_score:.2f}")
    
    if health_score > max_score:
        max_score = health_score
        strongest_cell_idx = i

# Result batayein
if strongest_cell_idx != -1:
    print(f"\nResult: Cell #{strongest_cell_idx + 1} is the STRONGEST cell.")
    print("Is cell ka life cycle aur growth potential sab se zyada hai.")

# 3. Screen par dikhayein
res_plotted = results[0].plot()
cv2.imshow("Strongest Cell Finder", res_plotted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import datetime
import os

# 1. Model Load karein
model = YOLO('best.pt')

# 2. Image Path
img_path = '20230304-17097-3-F15-157.jpg' 
img = cv2.imread(img_path)
results = model.predict(img_path, conf=0.5)

# 3. Cells ki Ginti aur Day Prediction Logic
boxes = results[0].boxes
count = len(boxes)

day_status = "Unknown"
if count == 1:
    day_status = "Day 1 (Zygote Stage)"
elif 2 <= count <= 4:
    day_status = "Day 2 (Cleavage Stage)"
elif 5 <= count <= 8:
    day_status = "Day 3 (Morula Shuru)"
elif count > 8:
    day_status = "Day 4+ (Advanced Stage)"

print(f"\n--- EMBRYO ANALYSIS REPORT ---")
print(f"Total Cells Detected: {count}")
print(f"Current Estimated Day: {day_status}")

# 4. Healthy Cell dhoondne ka Logic (Symmetry Check)
best_score = 0
best_box_idx = -1

for i, box in enumerate(boxes):
    x1, y1, x2, y2 = box.xyxy[0].tolist()
    w = x2 - x1
    h = y2 - y1
    symmetry = min(w, h) / max(w, h)
    health_score = (float(box.conf[0]) * 0.5) + (symmetry * 0.5)
    
    if health_score > best_score:
        best_score = health_score
        best_box_idx = i

# 5. Drawing (Green Box for Healthy, Red for Others)
for i, box in enumerate(boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
    
    if i == best_box_idx:
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 3)
        cv2.putText(img, "MOST HEALTHY", (x1, y1-10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
    else:
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

# --- SAVE KARNE WALA SECTION ---
# Folder ka naam
output_folder = 'Health_Analysis_Results'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Unique file name banana (Time aur Date ke sath)
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
save_filename = f"{output_folder}/Embryo_Health_{timestamp}.jpg"

# Image save karna
cv2.imwrite(save_filename, img)
print(f"Status: Analysis image saved as {save_filename}")
# -------------------------------

# Result Show Karein
cv2.imshow("Embryo Health Analysis", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import datetime
import os

# 1. Model Load karein
model = YOLO('best.pt')

# 2. Image load karein
img_path = 'D2024.06.02_S00699_I4778_P_WELL01_RUN248.jpg' 
img = cv2.imread(img_path)
results = model.predict(img_path, conf=0.5)

# 3. Basic Info nikaalein
boxes = results[0].boxes
count = len(boxes)

# Day Prediction Logic
if count == 1: day_status = "Day 1"
elif 2 <= count <= 4: day_status = "Day 2"
elif 5 <= count <= 8: day_status = "Day 3"
else: day_status = "Day 4+ (Morula)"

# 4. Healthy Cell Logic
best_score = 0
best_box_idx = -1
best_conf = 0

for i, box in enumerate(boxes):
    x1, y1, x2, y2 = box.xyxy[0].tolist()
    w, h = x2 - x1, y2 - y1
    symmetry = min(w, h) / max(w, h)
    conf = float(box.conf[0])
    health_score = (conf * 0.5) + (symmetry * 0.5)
    
    if health_score > best_score:
        best_score = health_score
        best_box_idx = i
        best_conf = conf

# 5. UI/Display: Image par details likhna
# Black sidebar background
cv2.rectangle(img, (0, 0), (280, 150), (0, 0, 0), -1) 

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, f"Total Cells: {count}", (10, 30), font, 0.7, (255, 255, 255), 2)
cv2.putText(img, f"Stage: {day_status}", (10, 60), font, 0.7, (255, 255, 255), 2)
cv2.putText(img, f"Healthy Cell Prob: {best_conf:.2f}", (10, 90), font, 0.7, (0, 255, 0), 2)
cv2.putText(img, "Green Box = Strongest", (10, 120), font, 0.6, (0, 255, 0), 1)

# 6. Boxes Draw karna
for i, box in enumerate(boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
    if i == best_box_idx:
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 3) # Green for Healthy
    else:
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 1) # Red for others

# --- SAVE KARNE WALA SECTION ---
# Folder banana agar nahi hai
output_dir = 'Embryo_Reports'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Time ke sath unique file name banana
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
save_path = f"{output_dir}/Report_{timestamp}.jpg"

cv2.imwrite(save_path, img)
print(f"Report saved successfully at: {save_path}")
# -------------------------------

# Show Result
cv2.imshow("Embryo Medical Report", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import datetime
import os

model = YOLO('best.pt')
img_path = '20230304-17097-3-F15-157.jpg' 
img = cv2.imread(img_path)
results = model.predict(img_path, conf=0.5)

boxes = results[0].boxes
count = len(boxes)
now = datetime.datetime.now()
current_time = now.strftime("%Y-%m-%d %H:%M")
file_timestamp = now.strftime("%Y%m%d_%H%M%S") # File name ke liye special format

# Advance Medical Logic
verdict = "Monitor Further"
color = (255, 255, 255) # White

if count >= 7 and results[0].boxes.conf.max() > 0.85:
    verdict = "High Growth Potential"
    color = (0, 255, 0) # Green
elif count < 4:
    verdict = "Slow Division"
    color = (0, 165, 255) # Orange

# Professional Dashboard Overlay
cv2.rectangle(img, (0, 0), (320, 200), (40, 40, 40), -1) 
cv2.rectangle(img, (0, 0), (320, 200), (200, 200, 200), 2) # Border

font = cv2.FONT_HERSHEY_DUPLEX
cv2.putText(img, f"EMBRYO REPORT", (15, 30), font, 0.7, (0, 255, 255), 2)
cv2.putText(img, f"Time: {current_time}", (15, 60), font, 0.5, (200, 200, 200), 1)
cv2.putText(img, f"Total Cells: {count}", (15, 90), font, 0.6, (255, 255, 255), 1)
cv2.putText(img, f"Verdict:", (15, 120), font, 0.6, (255, 255, 255), 1)
cv2.putText(img, verdict, (15, 150), font, 0.7, color, 2) 

# Healthy Cell Box Drawing
best_idx = results[0].boxes.conf.argmax()
for i, box in enumerate(boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
    if i == best_idx:
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 3) 
        cv2.putText(img, "SELECTED", (x1, y1-10), font, 0.5, (0, 255, 0), 2)
    else:
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 1)

# --- YAHAN SAY SAVE KARNE WALA CODE SHURU HAI ---
output_folder = 'Analysis_Results'
if not os.path.exists(output_folder):
    os.makedirs(output_folder) # Agar folder nahi hai to bana do

save_name = f"{output_folder}/Result_{file_timestamp}.jpg"
cv2.imwrite(save_name, img) # Image save ho rahi hai
print(f"--- SUCCESS: Report saved as {save_name} ---")
# -----------------------------------------------

cv2.imshow("Professional Embryo Analysis", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
